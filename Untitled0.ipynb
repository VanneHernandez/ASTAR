```python
# Importaciones
import matplotlib.pyplot as plt
import networkx as nx
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

# Definicion de la red Bayesiana
modelo = DiscreteBayesianNetwork([
    ('Nublado', 'Lluvia'),
    ('Nublado', 'Aspersor'),
    ('Lluvia', 'CespedMojado'),
    ('Aspersor', 'CespedMojado')
])

# Definición de Tablas de Probabilidad
# 0.7 = Probabilidad de que NO esté nublado
# 0.3 = Probabilidad de que SÍ esté nublado
tp_nublado = TabularCPD(variable='Nublado', variable_card=2, values=[[0.7], [0.3]])

tp_lluvia = TabularCPD(
    variable='Lluvia',
    variable_card=2,
    values=[[0.8, 0.2], [0.2, 0.8]],
    evidence=['Nublado'],
    evidence_card=[2]
)

tp_aspersor = TabularCPD(
    variable='Aspersor',
    variable_card=2,
    values=[[0.5, 0.9], [0.5, 0.1]],
    evidence=['Nublado'],
    evidence_card=[2]
)

tp_cesped = TabularCPD(
    variable='CespedMojado',
    variable_card=2,
    values=[[0.9, 0.6, 0.7, 0.01], [0.1, 0.4, 0.3, 0.99]],
    evidence=['Lluvia', 'Aspersor'],
    evidence_card=[2, 2]
)

modelo.add_cpds(tp_nublado, tp_lluvia, tp_aspersor, tp_cesped)

assert modelo.check_model()

# Visualizar la Red
G = nx.DiGraph()
G.add_nodes_from(modelo.nodes)
G.add_edges_from(modelo.edges)

plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True,
        node_color='skyblue',
        node_size=2000,
        font_size=12,
        font_weight='bold',
        arrows=True,
        arrowstyle='-|>',
        arrowsize=20)
plt.title("Red Bayesiana del Clima")
plt.show()

# Inferencia
Inferencia = VariableElimination(modelo)

resultado = Inferencia.query(variables=['Lluvia'], evidence={'CespedMojado': 1})
print("Probabilidad de que esté lloviendo dado que el cesped esté mojado:")
print(resultado)

resultado_nublado = Inferencia.query(variables =['CespedMojado'], evidence ={'Nublado' :1 })
print("\nProbabilidad de que el Cesped esté mojado si está nublado:")
print(resultado_nublado)

resultado_aspersor_encendido = Inferencia.query(variables=['Aspersor'], evidence={'CespedMojado': 1})
print("Probabilidad de que el Aspersor esté encendido dado que el cesped esté mojado:")
print(resultado_aspersor_encendido)
```
